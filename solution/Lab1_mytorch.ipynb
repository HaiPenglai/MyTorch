### 任务：简单了解pytorch包的文件结构

问题：`print(包.__file__)`打开的是什么文件？

【答案：安装后，这个包下面的`__init__.py`文件】

问题：通过pip install安装的pytorch和pip install -e安装的mytorch有什么区别

【答案：加上-e原地安装，否则安装到site-packages】

问题：mytorch安装的时候为什么要加e

【答案：因为修改可以立即生效，否则每次重新安装太麻烦】

问题：pytorch源码和pip install 后`site-packages`下的源码的区别

【答案：源码包含C++/CUDA实现，而`site-packages`仅包含 编译后的二进制文件和必要的 Python 模块】

问题：`torch.__version__`来自于哪里？

【答案：来自于torch\version.py。`torch\__init__.py`当中引入了__version__】
